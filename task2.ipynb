```python
import os
import tqdm
import torch
import pandas as pd
import numpy as np
from PIL import Image
import torch.nn as nn
import seaborn as sns
import tensorflow as tf
import torch.optim as optim
from matplotlib import pyplot as plt
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, roc_curve
from torch.utils.data import DataLoader, Dataset
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
(mnist_images, mnist_labels), (mnist_test_images, mnist_test_labels) = tf.keras.datasets.mnist.load_data()
```

```python
mnist_images = mnist_images / 255.0
mnist_test_images = mnist_test_images / 255.0
```

```python
mnist_images = torch.tensor(mnist_images, dtype=torch.float32)
mnist_labels = torch.tensor(mnist_labels, dtype=torch.long)
mnist_test_images = torch.tensor(mnist_test_images, dtype=torch.float32)
mnist_test_labels = torch.tensor(mnist_test_labels, dtype=torch.long)
```

```python
mnist_images = mnist_images.unsqueeze(1)
mnist_test_images = mnist_test_images.unsqueeze(1)
```

```python
transform = transforms.Compose([
    transforms.Normalize((0.5,), (0.5,))
])
```

```python
train_dataset = TensorDataset(mnist_images, mnist_labels)
test_dataset = TensorDataset(mnist_test_images, mnist_test_labels)
```

```python
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

```python
print(type(train_loader))
print(type(val_loader))
print(len(train_loader))
print(len(val_loader))
```

```python
train_features, train_labels = next(iter(train_loader))
print(f"BatchSize and after feature_shape: {train_features.size()}")
print(f"BatchSize and after lable_shape: {train_labels.size()}")
img = train_features[0].squeeze()
label = train_labels[0]
plt.imshow(img, cmap="gray")
plt.show()
print(f"Label: {label}")
```

```python
import torch.nn as nn
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()

        # Block 1
        self.conv1_1 = nn.Conv2d(in_channels=1, out_channels=64, kernel_size=3, padding=1)
        self.conv1_2 = nn.Conv2d(in_channels=64, out_channels=64, kernel_size=3, padding=1)
        self.maxpool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.dropout1_1 = nn.Dropout(p=0.2)

        # Block 2
        self.conv2_1 = nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding=1)
        self.conv2_2 = nn.Conv2d(in_channels=128, out_channels=128, kernel_size=3, padding=1)
        self.maxpool2 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.dropout2_1 = nn.Dropout(p=0.2)

        self.relu = nn.ReLU(inplace=True)
        self.flatten = nn.Flatten()

        self.fully_connected_1 = nn.Linear(in_features=128 * 7 * 7, out_features=120)
        self.dropout_fc = nn.Dropout(p=0.2)
        self.fully_connected_2 = nn.Linear(in_features=120, out_features=64)
        self.fully_connected_3 = nn.Linear(in_features=64, out_features=10)

    def forward(self, input_shape):
        # Block 1
        out = self.conv1_1(input_shape)
        out = self.relu(out)
        out = self.conv1_2(out)
        out = self.relu(out)
        out = self.maxpool1(out)
        out = self.dropout1_1(out)

        # Block 2
        out = self.conv2_1(out)
        out = self.relu(out)
        out = self.conv2_2(out)
        out = self.relu(out)
        out = self.maxpool2(out)
        out = self.dropout2_1(out)

        out = self.flatten(out)

        out = self.fully_connected_1(out)
        out = self.relu(out)
        out = self.dropout_fc(out)
        out = self.fully_connected_2(out)
        out = self.relu(out)
        out = self.fully_connected_3(out)

        return out
```

```python
model = CNN()
print(model)
model.to(device)
```

```python
loss_func = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)
```

```python
def train(model, train_loader, val_loader, loss_func, optimizer, epochs, patience):
    training_acc = []
    testing_acc = []
    training_loss = []
    testing_loss = []
    best_val_loss = float('inf')
    epochs_since_improvement = 0

    for epoch in tqdm.tqdm(range(epochs)):
        train_loss = 0.0
        train_acc = 0.0

        model.train()
        print("########## epochs : ",epoch+1," ########### ")
        for i,(images, labels) in enumerate(train_loader):
            images = images.to(device)
            labels = labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = loss_func(outputs, labels)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
            predicted = torch.max(outputs, 1)[1]
            train_acc += torch.sum(torch.eq(predicted, labels).float()).item()
            if i % 25 == 24:  # print every 100 mini-batches
               print ('Complete no of Mini batche(100) in one batch: ',i+1)
        training_acc.append(train_acc / len(train_loader.dataset))
        training_loss.append(train_loss / len(train_loader))

        model.eval()

        test_loss = 0.0
        test_acc = 0.0
        with torch.no_grad():
            for images, labels in val_loader:
                images = images.to(device)
                labels = labels.to(device)
                outputs = model(images)
                loss = loss_func(outputs, labels)
                test_loss += loss.item()
                predicted = torch.max(outputs, 1)[1]
                test_acc += torch.sum(torch.eq(predicted, labels).float()).item()

        testing_acc.append(test_acc / len(val_loader.dataset))
        testing_loss.append(test_loss / len(val_loader))

        print('Epoch {}, Training Accuracy: {:.2f}%, Testing Accuracy: {:.2f}%, Training Loss: {:.2f}, Testing loss: {:.2f}'
              .format(epoch + 1, training_acc[-1]*100, testing_acc[-1]*100, training_loss[-1], testing_loss[-1]))

        if testing_loss[-1] < best_val_loss:
            best_val_loss = testing_loss[-1]
            epochs_since_improvement = 0
        else:
            epochs_since_improvement += 1
            if epochs_since_improvement >= patience:
                print("Validation loss hasn't improved in {} epochs, stopping training".format(patience))
                break

    return model , training_loss , testing_loss , training_acc , testing_acc
```

```python
model , training_loss , testing_loss , training_acc , testing_acc = train ( model ,  train_loader , val_loader , loss_func , optimizer , 10 , 5)
```

```python
def test(model, test_loader):
    model.eval()
    predictions = []
    with torch.no_grad():
        for images, _ in test_loader:
            images = images.to(device)
            outputs = model(images)
            predicted = torch.max(outputs, 1)[1]
            predictions.append(predicted)
    return predictions
```

```python
predictions= test(model, val_loader)
```

```python
y_true = []
y_pred = []
with torch.no_grad():
    for images, labels in val_loader:
        images = images.to(device)
        labels = labels.to(device)
        outputs = model(images)
        predicted = torch.max(outputs, 1)[1]
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(predicted.cpu().numpy())
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_true, y_pred, digits=4))
```

```python
cm = confusion_matrix(y_true, y_pred)
class_labels = np.arange(cm.shape[0])
correct_pred = np.diag(cm)
wrong_pred = np.sum(cm, axis=1) - correct_pred
fig, ax = plt.subplots()
ax.bar(class_labels, correct_pred, label='Correct')
ax.bar(class_labels, wrong_pred, bottom=correct_pred, label='Wrong')
ax.set_xticks(class_labels)
ax.set_xlabel('Class')
ax.set_ylabel('Number of Predictions')
ax.legend()
plt.show()
```
